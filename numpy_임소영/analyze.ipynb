```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
#year = 2008
data_set_list = []
def industry(year):
    file = f'data_{year}.csv'
    data = pd.read_csv(file, index_col = [0])
    data.astype('int')
    data['avg'] = data.mean(axis = 1).round(2)
    #data
    # '분산' 열 추가
    data['var'] = data.var(axis = 1).round(2)
    #data
    # 사용할 데이터 셀렉
    use_data = data['avg']
    # concat할 리스트에 추가
    data_set_list.append(use_data)
    
    return data
```

```python
result = pd.concat(data_set_list, axis =1)
result
```

#### 그래프 그리기

```python
A = list(range(2008, 2019))
result.columns = A
result
```

```python
def draw_graph(kind):
    #kind = 'A'
    data_for_graph = result.loc[kind]
    #data_for_graph
    
    min_data = round(data_for_graph.min(), 0)
    max_data = round(data_for_graph.max(), 0)
    
    #plt.figure(figsize=(10, 10))
    plt.plot(data_for_graph)
    plt.xlim(2008, 2019)
    plt.ylim(min_data-10000, max_data+10000)
    plt.xlabel('YEAR')
    plt.ylabel('USAGE')
    plt.grid()
    plt.title(f'{kind}그룹 산업')
    
    plt.show()
```
