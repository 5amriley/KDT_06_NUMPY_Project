```python
import pandas as pd
import numpy as np
```

```python
def clean_data_2(insert_file, select_month):
    global A_group
    global B_group
    global C_group
    global D_group
    global E_group
    global F_group
    global G_group
    global H_group
    global I_group
    global J_group
    global K_group
    global L_group
    global N_group

    # file = './industry_data/2008_industry.csv'
    file = insert_file
    # data_real = pd.read_csv(file, skiprows=3)
    data = pd.read_csv(file)
    data = data.fillna(0)
    select_column = data.columns[3:]
    for x in select_column:
        data[x] = data[x].astype(str).str.replace(',', '')

    data_real = data.iloc[3:]

    # 컬럼 카테고리화하기 ------------------------------------------------------------------
    column_list = []
    for i in range(data.shape[1]):
        result = data.iloc[1][i]
        # print(i)
        # print(result)
        column_list.append(result)

    month = column_list[0]
    time = column_list[1]
    year = column_list[2]

    A = column_list[3:6]
    B = column_list[6:10]
    C = column_list[10:34]
    D = column_list[34]
    E = column_list[35:39]
    F = column_list[39:41]
    G = column_list[41:44]
    H = column_list[44:48]
    I = column_list[48:50]
    J = column_list[50:56]
    K = column_list[56:59]
    L = column_list[59]
    N = column_list[60]
    ETC = column_list[61:]  # ETC는 제외

    # 필요한 데이터만 남기고 슬라이싱 -----------------------------------------------------------
    data_real.columns = column_list
    data_df = data_real.T
    result = data_df.iloc[:61]
    my_data = result.T
    my_data

    # 월별 데이터 선택 -----------------------------------------------------------------------
    # select_month = '1월'
    month_01 = my_data.groupby('월별').get_group(select_month)

    # 'sum' 행 생성 ------------------------------------------------------------------------
    sample = []
    for i in range(61):
        sample.append(0)

    month_01.loc['sum'] = sample

    # 형변환 --------------------------------------------------------------------------------
    month_01[month_01.columns[3:]] = month_01[month_01.columns[3:]].astype('int64')

    # 'sum'에 총합 계산하거 추가해서 원본에 넣기 --------------------------------------------------
    result = month_01.select_dtypes(include='int64').sum()

    sample_list = [0, 0, 0]
    for i in range(len(result)):
        sample_list.append(result[i])
    month_01.loc['sum'] = sample_list

    # 'sum'열만 출력 -------------------------------------------------------------------------
    value = month_01.loc['sum']

    # 해당년도 해당월 합꼐 데이터 묶음 끼리의 총합 계산하기  -----------------------------------------
    A_result = value[3:6].sum()
    B_result = value[6:10].sum()
    C_result = value[10:34].sum()
    D_result = value[34].sum()
    E_result = value[35:39].sum()
    F_result = value[39:41].sum()
    G_result = value[41:44].sum()
    H_result = value[44:48].sum()
    I_result = value[48:50].sum()
    J_result = value[50:56].sum()
    K_result = value[56:59].sum()
    L_result = value[59].sum()
    N_result = value[60].sum()

    # 카테고리별로 리스트에 담기 ---------------------------------------------------------------
    A_group.append(A_result)
    B_group.append(B_result)
    C_group.append(C_result)
    D_group.append(D_result)
    E_group.append(E_result)
    F_group.append(F_result)
    G_group.append(G_result)
    H_group.append(H_result)
    I_group.append(I_result)
    J_group.append(J_result)
    K_group.append(K_result)
    L_group.append(L_result)
    N_group.append(N_result)
    #print(f"{select_month} 정리 끝")

def print_list():
    global A_group
    global B_group
    global C_group
    global D_group
    global E_group
    global F_group
    global G_group
    global H_group
    global I_group
    global J_group
    global K_group
    global L_group
    global N_group
    
    list_for_print = [A_group, B_group, C_group, D_group, E_group, F_group, G_group, H_group, I_group, J_group, K_group, L_group, N_group]
    for list in list_for_print:
        print(list)
```

```python
year = 2008

file = f'./industry_data/{year}_industry.csv'
A_group = []
B_group = []
C_group = []
D_group = []
E_group = []
F_group = []
G_group = []
H_group = []
I_group = []
J_group = []
K_group = []
L_group = []
N_group = []

for i in range(1,13):
    word = f"{i}월"
    clean_data_2(file, word)
    
print(f"{year}년 파일")
month_list = []
for i in range(1, 13):
    month_list.append(i)
print(month_list)
print_list()

# ------ csv 파일 저장 -------------
list_for_print = [A_group, B_group, C_group, D_group, E_group, F_group, G_group, H_group, I_group, J_group, K_group, L_group, N_group]
list_for_csv = []
#list_for_csv.append(month_list)
for list in list_for_print:
    list_for_csv.append(list)
    
final_data = pd.DataFrame(list_for_csv, columns=month_list, index = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I' ,'J', 'K', 'L', 'N'])
final_data

final_data.to_csv('data_2008.csv')
```
