```python
custom_order = veggiesDF['시도별'].unique()     # 기존의 '시도별' 컬럼 정렬 순서를 저장
# 사용자 지정한 기준으로 순서가 정의된 category(범주형) 타입 생성, 할당
veggiesDF['시도별'] = pd.Categorical(veggiesDF['시도별'], categories=custom_order, ordered=True)
# 두 열을 기준으로 재정렬 (인덱스도 초기화)
veggiesDF = veggiesDF.sort_values(['시도별', '시점'], ascending=True, ignore_index=True)
veggiesDF.head()
```

```python
def plot_veggies(location, veggie):
    """
    지역명, 채소명 지정하면 해당 정보를 막대그래프로 그리는 함수
    :param location: 지역명
    :param veggie: 채소명
    :return: None
    """
    fig = plt.figure(figsize=(10, 6))
    ax = fig.add_subplot(111)
    
    years = veggiesDF['시점'].unique()    # 년도 저장
    
    location_mask = veggiesDF['시도별'] == location
    locationDF = veggiesDF[location_mask]
    ax.bar(locationDF['시점'], locationDF[f'{veggie}:생산량 (톤)'])
    
    ax.set_title(f'[{location if location != "계" else "전국"}] 연간 {veggie} 생산량', fontsize=15)
    ax.set_xlabel('연도')
    ax.set_ylabel('생산량 (톤)')
    
    plt.show()
```

```python
def plot_all_veggies(location, target=0, corr=0):
    """
    6가지 종류의 채소 생산량 그래프를 한 화면에 출력하는 함수
    :param location: 지역명
    :param target: 0:[총 생산면적], 1:[총 생산량], 2:[노지 면적], 3:[노지 면적당 생산량], 4:[노지 총 생산량], 5:[시설 면적], 6:[시설 면적당 생산량], 7:[시설 총 생산량]
    :return: 회귀직선 기울기 리스트
    """
    global veggie_list
    
    lc_dict = {'서울특별시':'서울', '부산광역시':'부산', '인천광역시':'인천', '대구광역시':'대구', '대전광역시':'대전', '광주광역시':'광주', '울산광역시':'울산'}
    
    if location not in location_list:
        print('잘못된 입력입니다.')
        return
    
    fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10),
                         sharex=True, sharey=False)
    
    title_list = ['생산면적', '총 생산량', '노지 면적', '노지 면적당 생산량', 
                  '노지 총 생산량', '시설 면적', '시설 면적당 생산량', '시설 총 생산량']
    fig.suptitle(f'[{location if location != "계" else "전국"}] 연간 채소 {title_list[target]}', fontsize=15)
    location_mask = veggiesDF['시도별'] == location
    locationDF = veggiesDF[location_mask]
    xs = np.array(range(1992, 2023))
    poly_fit_list = []                          # 각 채소 생산량 회귀선 기울기만 저장할 리스트
    target_col = None                           # 알아볼 정보에 해당하는 컬럼명
    
    for i, veggie in enumerate(veggie_list):
        
        if target == 0: target_col = f'{veggie}:면적 (ha)'
        elif target == 1: target_col = f'{veggie}:생산량 (톤)'
        elif target == 2: target_col = f'노지{veggie}:면적 (ha)'
        elif target == 3: target_col = f'노지{veggie}:10a당 생산량 (kg)'
        elif target == 4: target_col = f'노지{veggie}:생산량 (톤)'
        elif target == 5: target_col = f'시설{veggie}:면적 (ha)'
        elif target == 6: target_col = f'시설{veggie}:10a당 생산량 (kg)'
        elif target == 7: target_col = f'시설{veggie}:생산량 (톤)'
        
        veggie_mask = locationDF[target_col] != 0
        dataX = locationDF[veggie_mask]['시점'].unique()    # 년도 저장
        dataY = locationDF[veggie_mask][target_col]
        poly_fit = np.polyfit(dataX, dataY, 1)  # 1차 회귀식 구하기
        poly_1d = np.poly1d(poly_fit)           # <class 'numpy.poly1d'>
        ys = poly_1d(xs)                        # poly_id가 객체인데 이게 왜 되지???
        poly_fit_list.append(poly_fit[0])       # 각 채소 생산량 회귀선 기울기만 저장
        
        # 그리는 영역을 선택
        ax = axes[i//3 , i%3]
        ax.set_ylim([3, locationDF[target_col].max() * 1.05])
        
        # 타이틀을 부여
        ax.set_title(veggie, fontsize=12)
        # ax.set_title('상관계수: ', loc='right')
        
        # 상관계수
        if corr==1 and (location in lc_dict):
            corrDF = pd.DataFrame({'강수량':rainDF[f'{lc_dict[location]} 강수량(mm)'].reset_index(drop=True), 
                                   '데이터':veggiesDF[location_mask][target_col].reset_index(drop=True)})
            no_zero = corrDF['데이터'] != 0
            
            #veggie_rain_corr = np.corrcoef(rainDF[f'{lc_dict[location]} 강수량(mm)'], veggiesDF[region_mask][target_col], ddof=0)[0, 1]
            veggie_rain_corr = np.corrcoef(corrDF[no_zero]['강수량'], corrDF[no_zero]['데이터'], ddof=0)[0, 1]
            ax.set_title(f'상관계수: {veggie_rain_corr:.3f}', loc='right', color='red')
            #ax.scatter(rainDF[f'{lc_dict[location]} 강수량(mm)'], veggiesDF[region_mask][target_col])
            ax.scatter(corrDF[no_zero]['강수량'], corrDF[no_zero]['데이터'], c='g', s=30)
            ax.set_xlabel('강수량 (mm)',fontsize=12)
            ax.set_ylabel('면적당 생산량',fontsize=12)
            #print(list(zip(rainDF[f'{lc_dict[location]} 강수량(mm)'], veggiesDF[region_mask][target_col])))
        
        if corr==2 and (location in lc_dict):
            corrDF = pd.DataFrame({'기온':tempDF[f'{lc_dict[location]} 평균기온(℃)'].reset_index(drop=True), 
                                   '데이터':veggiesDF[location_mask][target_col].reset_index(drop=True)})
            no_zero = corrDF['데이터'] != 0
            
            veggie_rain_corr = np.corrcoef(corrDF[no_zero]['기온'], corrDF[no_zero]['데이터'], ddof=0)[0, 1]
            ax.set_title(f'상관계수: {veggie_rain_corr:.3f}', loc='right', color='red')
            ax.scatter(corrDF[no_zero]['기온'], corrDF[no_zero]['데이터'], c='g', s=30)
            ax.set_xlabel('평균기온(℃)',fontsize=12)
            ax.set_ylabel('면적당 생산량',fontsize=12)
        
        if corr == 0:
            ax.plot(xs, ys, color='red')
            ax.bar(locationDF['시점'], locationDF[target_col])
        
    
    # 그래프 사이의 간격을 좁힌다.
    plt.tight_layout()
    plt.savefig(f'./pics/{location if location != "계" else "전국"}.png', format='png', dpi=300)
    plt.show()
    
    return poly_fit_list
```
